## ipython notebookで行うPythonによるFIRフィルタの設計

サンプリング周波数を624kHzから312kHzにダウンサンプルするためのFIRフィルタを設計します。フィルタは624kHzで設計します。間引きは実行時に行います。

グラフをインラインに表示するよう設定。綺麗な表示になるようsvgを指定(オプション)

```python
%matplotlib inline
%config InlineBackend.figure_format='svg'
```

必要なパッケージをインポートする。以後np, plという識別子は、numpyとpylabへの参照です。

```python
import numpy as np
import pylab as pl
from scipy import signal
```

LPFをfirwin2関数で設計する。32タップでカットオフ周波数fc=156kHz, サンプリング周波数624kHz、すなわちナイキスト周波数312kHzの条件で、目標ゲインとその周波数を配列で与えている。

```python
freq = [0, 156e3, 156e3, 312e3]
gain = [1, 1, 0, 0]
nyq = 312e3
taps = signal.firwin2(32, freq, gain, nyq=nyq)
```

得られたタップ係数を確認してみる

```python
taps
```

周波数特性をfreqz関数で計算、そしてグラフにプロットする。横軸は周波数に、縦軸はdBに換算している。

```python
ww,hh = signal.freqz(taps)
pl.grid()
pl.xlabel('Frequency (Hz)')
pl.ylabel('Gain (dB)')
pl.xlim((0,nyq))
pl.ylim((-80,0))
pl.plot(ww * nyq / np.pi, 20*np.log10(hh))
```

タップ係数を16ビット固定少数に変換する。これをソースコードにそのままコピーして使うことができる。

```python
np.vectorize(int)(taps * 32768)
```

タップ係数、すなわちインパルス応答をプロットしてみる

```python
pl.grid()
pl.plot(taps)
pl.savefig('FigF-ImpulseResponse.png', dpi=150)
```

以上、簡単なFIRフィルタの設計例を示しました。実際にはタップ数やカットオフ周波数などのパラメータを変えながら繰り返し実行して望む特性を確認しながら設計を行っています。
